```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Se procede a cargar los datos de la data obtenida en el proyecto de NASA API, en dicho notebook se puede ver todo el procedimiento para la extracción y transformación de los datos. 

```python
ps_comp = pd.read_csv('ps_comp_data.csv')
```

#### Documentación de Planetary systems composite

https://exoplanetarchive.ipac.caltech.edu/docs/API_PS_columns.html

Descripción de cada una de las columnas:

- **planet_name:** Planet Name
- **hostname:** Stellar name 
- **sys_num_stars:** Number of Stars
- **sys_num_planet:** Number of Planets 
- **discovery_method:** Discovery Method
- **discovery_year:** Discovery Year
- **discovery_telescope:** Discovery Telescope
- **orbital_period:** Orbital Period, unit=days 
- **orbital_semi_major:** Orbit Semi-Major Axis, unit = au 
- **planet_radius_earth:** Planet Radius (Earth Radius)
- **planet_radius_jupiter:** Planet Radius (Jupiter Radius) 
- **planet_mass_earth:** Planet Mass (Earth Mass)
- **planet_mass_jupiter:** Planet Mass (Jupiter Mass) 
- **eccentricity:** Eccentricity
- **stellar_eff_temp:** Stellar Effective Temperature unit=K 
- **stellar_radius:** Stellar Radius (Solar Radius) 
- **stellar_mass:** Stellar Mass
- **dist_system:** Distance, unit = pc

## Orbital period vs orbital semi major

La relación entre el periodo orbital y el semi eje mayor es de esperarse, ya que al tener un semi eje mayor más elevado el exoplaneta va a estar más alejado de su estrella por lo tanto la atracción gravitacional no es tan intensa lo que genera que está vaya más lento o lo que es igual que le tome mucho más tiempo el dar una vuelta alrededor de su estrella. 


Se tiene un semieje mayor y un semieje menor, ya que las orbitas planetarias son elipses, con diferentes grados de excentricidad. 


Esto puede ser entendido de mejor manera a través de las Leyes de Kepler, en el siguiente enlace puede encontrar una descripción en detalle sobre estas leyes y otros puntos de interés:

https://www.fisicalab.com/apartado/leyes-kepler

```python
sns.set(context='paper', style='darkgrid', rc={'figure.figsize': (15, 10)})


ax = sns.scatterplot(x='orbital_period', y='orbital_semi_major', size='planet_radius_jupiter', data=ps_comp, 
                     hue='discovery_method', alpha=0.6)
ax.set_title('Orbital Semi major vs Orbital Period', fontsize=30)

ax.set_xscale('log')
ax.set_yscale('log')
```

```python
ax = sns.histplot(ps_comp['planet_radius_jupiter'])

ax.set_title('Distribución por Radio del Planeta', fontsize=30);
```

### Filtrado

Se hace un filtrado de ciertos exoplanetas extremos para poder observar mejor el comportamiento de la muestra. 

```python
ps_comp_copy = ps_comp.copy()
```

```python
ps_comp_copy
```

```python
ps_comp_copy = ps_comp_copy.loc[ps_comp_copy['orbital_period'] < 1000000]
```

```python
sns.set(context='paper', style='darkgrid', rc={'figure.figsize': (15, 10)})


ax = sns.scatterplot(x='orbital_period', y='orbital_semi_major', size='planet_radius_jupiter', data=ps_comp_copy, 
                     hue='discovery_method', alpha=0.6)

ax.set_xscale('log')
ax.set_yscale('log')
```

Se puede observar que el método de descubrimiento destacado dentro del dataset es el método del tránsito.
